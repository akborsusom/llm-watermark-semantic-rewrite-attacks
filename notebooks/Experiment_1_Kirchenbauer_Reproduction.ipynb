```python
# Strong rewrite attack using a local paraphrase model
!pip install -q sentencepiece sacremoses
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

para_model_name = "Vamsi/T5_Paraphrase_Paws"
para_tokenizer = AutoTokenizer.from_pretrained(para_model_name)
para_model = AutoModelForSeq2SeqLM.from_pretrained(para_model_name).to(device)

def paraphrase_text(text, seed=123):
    prompt = "paraphrase: " + text + " </s>"
    inputs = para_tokenizer(prompt, return_tensors="pt", truncation=True, max_length=512).to(device)

    torch.manual_seed(seed)
    outputs = para_model.generate(
        **inputs,
        max_length=256,
        do_sample=True,
        top_k=120,
        top_p=0.95,
        temperature=1.4,
        num_return_sequences=1
    )
    return para_tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```
