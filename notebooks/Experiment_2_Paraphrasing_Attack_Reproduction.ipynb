```python
import pandas as pd

df = pd.read_excel("/baseline_379_results.xlsx")

print("Rows:", len(df))
df.head()
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)

para_model_name = "Vamsi/T5_Paraphrase_Paws"

para_tokenizer = AutoTokenizer.from_pretrained(para_model_name)
para_model = AutoModelForSeq2SeqLM.from_pretrained(para_model_name).to(device)
para_model.eval()

print("Paraphrase model loaded.")
```

```python
def paraphrase_text(text, seed=123):
    prompt = "paraphrase: " + str(text) + " </s>"

    inputs = para_tokenizer(
        prompt,
        return_tensors="pt",
        truncation=True,
        max_length=512
    ).to(device)

    torch.manual_seed(seed)

    outputs = para_model.generate(
        **inputs,
        max_length=256,
        do_sample=True,
        top_k=120,
        top_p=0.95,
        temperature=1.4,
        num_return_sequences=1
    )

    return para_tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
sample_text = df.loc[0, "watermarked_text"]

print("ORIGINAL:")
print(sample_text[:1000])

print("\nPARAPHRASED:")
print(paraphrase_text(sample_text, seed=999))
```

```python
from transformers import AutoTokenizer
from extended_watermark_processor import WatermarkDetector

detector_tokenizer = AutoTokenizer.from_pretrained("gpt2")
vocab = list(detector_tokenizer.get_vocab().values())

gamma = 0.25
z_threshold = 4.0

detector = WatermarkDetector(
    vocab=vocab,
    gamma=gamma,
    seeding_scheme="selfhash",
    device=device,
    tokenizer=detector_tokenizer,
    z_threshold=z_threshold,
    normalizers=[],
    ignore_repeated_ngrams=False,
    select_green_tokens=True
)

print("Detector ready.")
```

```python
original_detection = detector.detect(sample_text)
paraphrased = paraphrase_text(sample_text, seed=999)
paraphrase_detection = detector.detect(paraphrased)

print("Before z-score:", original_detection["z_score"])
print("Before prediction:", original_detection["prediction"])

print("After z-score:", paraphrase_detection["z_score"])
print("After prediction:", paraphrase_detection["prediction"])
```

```python
attack_results = []

for i in range(len(df)):

    if i % 10 == 0:
        print(f"Running {i+1}/{len(df)}")

    original = df.loc[i, "watermarked_text"]

    before = detector.detect(original)

    paraphrased = paraphrase_text(original, seed=123+i)

    after = detector.detect(paraphrased)

    attack_results.append({
        "id": i,
        "prompt": df.loc[i, "prompt"],
        "before_z_score": before["z_score"],
        "after_z_score": after["z_score"],
        "z_score_drop": before["z_score"]-after["z_score"],
        "before_prediction": before["prediction"],
        "after_prediction": after["prediction"],
        "original_text": original,
        "paraphrased_text": paraphrased
    })
```

```python
attack_df = pd.DataFrame(attack_results)

print("Before detection rate:", attack_df["before_prediction"].mean())
print("After detection rate:", attack_df["after_prediction"].mean())

print("Average z-score before:", attack_df["before_z_score"].mean())
print("Average z-score after:", attack_df["after_z_score"].mean())

print("Average z-score drop:", attack_df["z_score_drop"].mean())
```

```python
attack_df.to_excel("experiment2_paraphrasing_results.xlsx", index=False)
```
